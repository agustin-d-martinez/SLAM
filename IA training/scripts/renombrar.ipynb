```python
import os

def renombrar_archivos_numerados(carpeta, inicio=1):
    # Obtener lista de archivos (solo archivos, no carpetas)
    archivos = [
        f for f in os.listdir(carpeta)
        if os.path.isfile(os.path.join(carpeta, f))
    ]
    
    # Ordenar por nombre (en tu caso equivale a orden cronológico)
    archivos.sort()

    contador = inicio

    for archivo in archivos:
        nombre, extension = os.path.splitext(archivo)

        nuevo_nombre = f"{contador}{extension}"
        ruta_vieja = os.path.join(carpeta, archivo)
        ruta_nueva = os.path.join(carpeta, nuevo_nombre)

        os.rename(ruta_vieja, ruta_nueva)
        contador += 1
```
