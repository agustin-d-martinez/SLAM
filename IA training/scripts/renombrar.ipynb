```python
from pathlib import Path
from PIL import Image, ImageOps
```

```python
def renombrar_archivos(carpeta, inicio=1):
    carpeta = Path(carpeta)

    # Obtener todos los archivos
    archivos = [f for f in carpeta.iterdir() if f.is_file()]

    # Ordenar por nombre (YYYYMMDD_HHMMSS)
    archivos.sort(key=lambda x: x.stem)

    # Renombrado temporal para evitar conflictos
    temporales = []

    for i, archivo in enumerate(archivos):
        temp = carpeta / f"__temp__{i}{archivo.suffix}"
        archivo.rename(temp)
        temporales.append(temp)

    # Renombrado final
    for numero, archivo in enumerate(temporales, start=inicio):
        nuevo = carpeta / f"{numero}{archivo.suffix}"
        archivo.rename(nuevo)

    print(f"Renombrados {len(temporales)} archivos.")

def rotar_imagenes(carpeta, grados):

    if grados not in (90, 180, 270):
        raise ValueError("Los grados deben ser 90, 180 o 270")

    carpeta = Path(carpeta)

    extensiones = {".jpg", ".jpeg", ".png", ".webp", ".bmp", ".tiff"}

    archivos = [
        f for f in carpeta.iterdir()
        if f.is_file() and f.suffix.lower() in extensiones
    ]

    for archivo in archivos:
        try:
            with Image.open(archivo) as img:

                # Aplicar orientación EXIF real
                img = ImageOps.exif_transpose(img)

                # Rotación horaria
                img_rotada = img.rotate(-grados, expand=True)

                # Guardar sobrescribiendo
                img_rotada.save(
                    archivo,
                    quality=100,
                    subsampling=0
                )

            print(f"Rotada: {archivo.name}")

        except Exception as e:
            print(f"Error con {archivo.name}: {e}")
```
